```python
from genetic_selection import GeneticSelectionCV
```

```python

import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
```

```python
df_training = pd.read_csv("/content/drive/MyDrive/Genetic Algorithm/data.csv")
```

```python
df_training.shape
```

```python
df_training.head()
```

```python
df_training.isnull().sum()
```

```python
df_training['Unnamed: 32'].fillna(0, inplace = True)
df_training.head()
```

```python
df_training.isnull().sum()
```

#Dividing into Features and Class

```python
X = df_training.drop('diagnosis', axis=1).copy()
X.head()
```

```python
y = df_training['diagnosis'].copy()
y.head()
```

```python
print(X.shape)
print(y.shape)
```

#Feature Selection and Modeling

```python
#estimator = DecisionTreeClassifier()
estimator = RandomForestClassifier(max_depth=6, random_state=0)
model = GeneticSelectionCV(
    estimator, cv=5, verbose=0,
    scoring="accuracy", max_features=10,
    n_population=100, crossover_proba=0.5,
    mutation_proba=0.2, n_generations=50,
    crossover_independent_proba=0.5,
    mutation_independent_proba=0.04,
    tournament_size=3, n_gen_no_change=10,
    caching=True, n_jobs=-1)

model = model.fit(X, y)
print('Features:', X.columns[model.support_])
```

#Memory Reduction

```python
for column in df1.columns:
    if df1[column].dtype == "float64":
        df1[column]=pd.to_numeric(df[column], downcast="float")
    if df1[column].dtype == "int64":
        df1[column]=pd.to_numeric(df[column], downcast="integer")
```

```python
df1.head()
```

#Dividing into Features and Class

```python
X = df1.drop('alive', axis=1).copy()
X.head()
```

```python
y = df1['alive'].copy()
y.head()
```

```python
print(X.shape)
print(y.shape)
```

#Feature Selection and Modeling

```python
#estimator = DecisionTreeClassifier()
estimator = RandomForestClassifier(max_depth=6, random_state=0)
model = GeneticSelectionCV(
    estimator, cv=5, verbose=0,
    scoring="accuracy", max_features=7,
    n_population=100, crossover_proba=0.5,
    mutation_proba=0.2, n_generations=50,
    crossover_independent_proba=0.5,
    mutation_independent_proba=0.04,
    tournament_size=3, n_gen_no_change=10,
    caching=True, n_jobs=-1)

model = model.fit(X, y)
print('Features:', X.columns[model.support_])
```
